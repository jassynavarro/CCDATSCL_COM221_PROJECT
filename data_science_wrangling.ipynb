# C. Cleaning

```python
df["date_completed"] = df["date_completed"].fillna(pd.NaT)
df['task_name'] = df['task_name'].fillna('None')
df['category'] = df['category'].fillna('None')
df['status'] = df['status'].fillna('None')
df['work_alone'] = df['work_alone'].fillna('None')
df['got_distracted'] = df['got_distracted'].fillna('None')
df['when_worked'] = df['when_worked'].fillna('None')
df['notes'] = df['notes'].fillna('None')
```

# D. Enriching

```python
df.dtypes
```

```python
df["date_created"] = pd.to_datetime(df["date_created"])
df["date_completed"] = pd.to_datetime(df["date_completed"])

df["completion_days"] = (
    df["date_completed"] - df["date_created"]
).dt.days
```

```python
df["created_weekday"] = df["date_created"].dt.day_name()
```

```python
df["created_weekday"] = df["date_created"].dt.day_name()
df["is_weekend"] = df["date_created"].dt.dayofweek >= 5  # True/False
df["is_weekend"] = df["is_weekend"].astype(int)  # 1 = weekend, 0 = weekday
```

```python
# Simple yes/no to 0/1
df["work_alone_num"] = df["work_alone"].map({"yes": 1, "no": 0})

# For got_distracted, assign numeric values
df["got_distracted_num"] = df["got_distracted"].map({
    "no": 0,
    "yes-little": 1,
    "yes-alot": 2
})
```

```python
df["day_quality_num"] = df["day_quality"].map({
    "unproductive": 0,
    "normal": 1,
    "productive": 2
})
```

```python
df["work_alone_num"] = df["work_alone_num"].fillna('NaN')
df["got_distracted_num"] = df["got_distracted_num"].fillna('NaN')
df["completion_days"] = df["completion_days"].fillna('NaN')
```

```python
# Clean and convert completion_days properly
df['completion_days'] = (
    df['completion_days']
    .astype(str).str.strip()
    .replace({'': None, 'nan': None, 'NaN': None, 'None': None})
)

df['completion_days'] = pd.to_numeric(df['completion_days'], errors='coerce')
```

```python
for col in ['got_distracted_num', 'work_alone_num', 'day_quality_num']:
    df[col] = (
        df[col].astype(str).str.strip()
        .replace({'': None, 'nan': None, 'NaN': None, 'None': None})
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# 1️⃣ If status == 'done', date_completed should not be NaT
invalid_done_dates = df[(df['status'] == 'done') & (df['date_completed'].isna())]
print("Invalid done with missing completion date:", len(invalid_done_dates))
display(invalid_done_dates)


# 2️⃣ If status != 'done', completion_days should be NaN
invalid_completion_days = df[(df['status'] != 'done') & (df['completion_days'].notna())]
print("Invalid non-done with completion_days:", len(invalid_completion_days))
display(invalid_completion_days)


# 3️⃣ completion_days should not be negative
df['completion_days'] = pd.to_numeric(df['completion_days'], errors='coerce')

negative_completion = df[df['completion_days'] < 0]
print("Negative completion_days:", len(negative_completion))
display(negative_completion)


# 4️⃣ day_quality_num must be within {0,1,2}
invalid_day_quality = df[
    df['day_quality_num'].notna() &
    ~df['day_quality_num'].isin([0,1,2])
]
print("Invalid day_quality_num values:", len(invalid_day_quality))
display(invalid_day_quality)


# 5️⃣ got_distracted_num must be within {0,1,2}
invalid_distraction = df[
    df['got_distracted_num'].notna() &
    ~df['got_distracted_num'].isin([0,1,2])
]
print("Invalid got_distracted_num values:", len(invalid_distraction))
display(invalid_distraction)


# 6️⃣ work_alone_num must be within {0,1}
invalid_work_alone = df[
    df['work_alone_num'].notna() &
    ~df['work_alone_num'].isin([0,1])
]
print("Invalid work_alone_num values:", len(invalid_work_alone))
display(invalid_work_alone)
```
