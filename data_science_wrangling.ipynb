```python
df["date_completed"] = pd.to_datetime(df["date_completed"], errors="coerce").dt.normalize()
```

# C. Cleaning

```python
df
```

```python
df.isna().sum()
```

```python
df["date_completed"] = df["date_completed"].fillna(pd.NaT)
df['task_name'] = df['task_name'].fillna('None')
df['category'] = df['category'].fillna('None')
df['status'] = df['status'].fillna('None')
df['work_alone'] = df['work_alone'].fillna('None')
df['got_distracted'] = df['got_distracted'].fillna('None')
df['when_worked'] = df['when_worked'].fillna('None')
df['notes'] = df['notes'].fillna('None')
```

```python
df.isna().sum()
```

```python
df
```

# D. Enriching

```python
df.dtypes
```

```python
df["date_created"] = pd.to_datetime(df["date_created"])
df["date_completed"] = pd.to_datetime(df["date_completed"])

df["completion_days"] = (
    df["date_completed"] - df["date_created"]
).dt.days
```

```python
df["created_weekday"] = df["date_created"].dt.day_name()
```

```python
df["created_weekday"] = df["date_created"].dt.day_name()
df["is_weekend"] = df["date_created"].dt.dayofweek >= 5  # True/False
df["is_weekend"] = df["is_weekend"].astype(int)  # 1 = weekend, 0 = weekday
```

```python
# Simple yes/no to 0/1
df["work_alone_num"] = df["work_alone"].map({"yes": 1, "no": 0})

# For got_distracted, assign numeric values
df["got_distracted_num"] = df["got_distracted"].map({
    "no": 0,
    "yes-little": 1,
    "yes-alot": 2
})
```

```python
df["day_quality_num"] = df["day_quality"].map({
    "unproductive": 0,
    "normal": 1,
    "productive": 2
})
```

# F. Publishing

```python
final_columns = [
"date_created",
"date_completed",
"task_id",
"task_name",
"category",
"status",
"when_worked",
"notes",
"completion_days",
"created_weekday",
"is_weekend",
"work_alone_num",
"got_distracted_num",
"day_quality_num"
]

df_final = df[final_columns]

# Display the cleaned DataFrame for publication
print(df_final.head())
```
